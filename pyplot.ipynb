Question 2

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
import matplotlib.pyplot as plt
import random

# Generate a sequence of at least 5 random numbers
random_numbers = [random.randint(1, 100) for _ in range(5)]

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Visualize the data using scatter plot
ax1.scatter(range(len(random_numbers)), random_numbers, color='red')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Numbers')

# Visualize the data using line plot (plot)
ax2.plot(range(len(random_numbers)), random_numbers, color='blue')
ax2.set_title('Line Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Numbers')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import matplotlib.pyplot as plt
import random

# Generate a sequence of at least 5 random numbers
random_numbers = [random.randint(1, 100) for _ in range(5)]

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 5), facecolor='lavender')

# Visualize the data using scatter plot
ax1.scatter(range(len(random_numbers)), random_numbers, color='deeppink', marker='^')
ax1.set_title('Scatter Plot of Random Numbers')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')
ax1.set_facecolor('lavender') #Change background color to lavender 

# Visualize the data using line plot (plot)
ax2.plot(range(len(random_numbers)), random_numbers, color='deeppink', linestyle=':')
ax2.set_title('Line Plot of Random Numbers')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')
ax2.set_facecolor('lavender') #Change background color to lavender 

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

4. Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

```python
import matplotlib.pyplot as plt

# Given data
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Plotting both datasets on the same set of axes
plt.figure(figsize=(10, 6))

# Plot insertion sort data
plt.plot(sizes, insertion_sort_times, marker='o', linestyle='-', color='purple', label='Insertion Sort')

# Plot merge sort data
plt.plot(sizes, merge_sort_times, marker='s', linestyle='-', color='deeppink', label='Merge Sort')

# Add labels and title
plt.xlabel('Size')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Performance')
plt.xscale('log')  # Set logarithmic scale for better visualization of large range of sizes
plt.yscale('log')  # Set logarithmic scale for better visualization of large range of times

# Add legend
plt.legend()

# Show plot
plt.show()
```
